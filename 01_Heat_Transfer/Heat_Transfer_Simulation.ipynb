dT/dt = (Q_in – hA (T – T_env)) / (m Cp)

Q_in = 200 W
hA = 15 W/K
T_env = 25°C

T0 = 20°C
0–1000 s

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

Q_in = 200
hA = 15
T_env = 25
T0 = 20
m = 1
Cp = 4184

time_span = (0, 1000)
time_eval = np.linspace(0, 1000, 500)

def dTdt(t, T):
    return (Q_in - hA*(T - T_env)) / (m*Cp)

y0 = [T0]

sol = solve_ivp(dTdt, time_span, y0, t_eval=time_eval)

plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```
